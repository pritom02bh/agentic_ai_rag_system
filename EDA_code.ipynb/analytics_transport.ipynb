```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'E:\agentic_ai_rag_system\data\transport_history.csv', encoding='ISO-8859-1')
```

```python
data.head()
```

```python
# Extract relevant columns
ports_data = data[['OriginPort', 'OriginPortCode', 'OriginCountry', 'DestinationPort', 'DestinationPortCode', 'DestinationCountry']]

# Create a DataFrame for Origin Ports
origin_ports = ports_data[['OriginPort', 'OriginPortCode', 'OriginCountry']].dropna().drop_duplicates()
origin_ports.columns = ['Port', 'PortCode', 'Country']

# Create a DataFrame for Destination Ports
destination_ports = ports_data[['DestinationPort', 'DestinationPortCode', 'DestinationCountry']].dropna().drop_duplicates()
destination_ports.columns = ['Port', 'PortCode', 'Country']

# Concatenate both DataFrames and drop duplicates
unique_ports = pd.concat([origin_ports, destination_ports]).drop_duplicates().reset_index(drop=True)

# Display the unique ports
unique_ports
```
